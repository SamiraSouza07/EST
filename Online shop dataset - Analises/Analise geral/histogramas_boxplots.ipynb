```python
import pandas as pd
shopdataset_comNull = pd.read_excel('./basededados.xlsx')
# Removendo os valores null
shopdataset = shopdataset_comNull.dropna()

```

```python
print(shopdataset.isnull().sum())
```

```python
print(shopdataset.notnull().sum())
```

<center>Discount_pct</center>

```python
#Calculando a média do Discount_pct
mediaDiscount_pct = round(shopdataset['Discount_pct'].mean(),2)
print('Média discount pct: ', mediaDiscount_pct)
```

```python
#Calculando a amplitude do Discount_pct
amplitudeDiscount_pct = shopdataset['Discount_pct'].max() - shopdataset['Discount_pct'].min()
print('Amplitude Discount_pct', amplitudeDiscount_pct)
```

```python
#Calculando a moda do Discount_pct
modaDiscount_pct = shopdataset['Discount_pct'].mode()
print('moda Discount_pct', modaDiscount_pct[0])
```

```python
#Calculando o DAM Discount_pct
shopdataset['Desvio_Absoluto'] = abs(mediaDiscount_pct-shopdataset['Discount_pct'])
damDiscount_pct = round(shopdataset['Desvio_Absoluto'].mean(),2)
display(damDiscount_pct)
```

```python
#Calculando a moda do Offline_Spend
modaOffline_Spend = shopdataset['Offline_Spend'].mode()
print('moda Offline Spend', modaOffline_Spend[0])
```

```python
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt
```
