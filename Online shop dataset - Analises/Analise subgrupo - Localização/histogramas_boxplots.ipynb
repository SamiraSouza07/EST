# Boxplot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import numpy as np
```

```python
shopdataset_comNull = pd.read_excel('./basededados.xlsx')
# Removendo os valores null
shopdataset = shopdataset_comNull.dropna()

```

```python
sns.boxplot(shopdataset, y='Location',x='Avg_Price', showfliers=True, flierprops=dict(marker='o', markersize=5, markerfacecolor='#EFA222'),showmeans=True,color='#00CCFF')
plt.xlabel('Preo da venda')
plt.title('Boxplot dos preos de venda')
sns.stripplot(shopdataset, y='Location',x='Avg_Price', color='#5022A6', alpha=0.2, size=1)
plt.show()

```

```python
sns.boxplot(shopdataset, y='Location',x='Online_Spend', showfliers=True, flierprops=dict(marker='o', markersize=5, markerfacecolor='#EFA222'),showmeans=True,color='#00CCFF')
plt.xlabel('Preo das vendas online')
plt.title('Boxplot dos preos de venda online')
sns.stripplot(shopdataset, y='Location',x='Online_Spend', color='#5022A6', alpha=0.2, size=1, )
plt.show()
```

```python
sns.boxplot(shopdataset, y='Location',x='Offline_Spend', showfliers=True, flierprops=dict(marker='o', markersize=5, markerfacecolor='#EFA222'),showmeans=True,color='#00CCFF')
plt.xlabel('Preo das vendas Offline')
plt.title('Boxplot dos preos de venda Offline')
sns.stripplot(shopdataset, y='Location',x='Offline_Spend', color='#5022A6', alpha=0.2, size=1)
plt.show()
```

```python
sns.boxplot(shopdataset, y='Location',x='Discount_pct', showfliers=True, flierprops=dict(marker='o', markersize=5, markerfacecolor='#EFA222'),showmeans=True,color='#00CCFF')
plt.xlabel('Percentual de desconto')
plt.title('Boxplot dos percentuais de desconto')
sns.stripplot(shopdataset, y='Location',x='Discount_pct', color='#5022A6', alpha=0.2, size=1)
plt.show()
```
